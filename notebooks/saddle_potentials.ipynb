```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Plot potentials with interesting saddles

```python
# Define the range for w1 and w2
wx = np.linspace(-2, 2, 100)
wy = np.linspace(-2, 2, 100)

# Create a meshgrid for w1 and w2
w1, w2 = np.meshgrid(wx, wy)

# Calculate the potential V(w1, w2)
v = (w1**2 - w2)**2(-w1**2 - w2)**2

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(w1, w2, v, cmap='viridis', edgecolor='none')

# Add labels and title
ax.set_xlabel('$w_1$')
ax.set_ylabel('$w_2$')
ax.set_zlabel('$V(w_1, w_2)$')
ax.set_title('Plot of the potential $V(w_1, w_2) =  w1**2 - w2 * w1**2 + 1/4 * w1**4$')

# Add a color bar
fig.colorbar(surface)

plt.show()
```
